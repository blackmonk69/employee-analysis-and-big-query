```python
df= pd.read_csv ("employee_dataset.csv")
```

```python
df.info()
```

we generate de bins for salary while the bins for age will be defined inside looker studio

```python
bins=[0,50000,150000,300000,500000,1000000]
group_names = ["Under 50K","50K-150K","150K-300K","300K-500K","Above 500K"]
df["bin_salary"]=pd.cut(df["Salary"],bins,labels=group_names)
```

```python
df.head()
```

let's get a column for full name

```python
df["full name"]=df["FirstName"] + " " + df["LastName"]
```

```python
df2=pd.read_csv("EducationLevel.csv")
```

we can merge the 2 tables to get the column education level

```python
df3=pd.merge(df,df2,left_on=["Education"],right_on=["EducationLevelID"])
```
